```python
import matplotlib.pyplot as plt
```

```python
#elbow method
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(data_stock)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,15),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Identify which all stocks are moving together and which all stocks are different from each other. 

```python
data_cor= data_stock.corr()
```

```python
data_cor.head()
```

```python
print("Stocks moving together")
for i in data_cor.columns:
    print("*"*30)
    print(i)
    print("*"*30)
    x= str(i)
    a=data_cor[data_cor["{}".format(x)]>0.9]
    
    print(a.index)
    print("=="*40)
```

```python
print("stocks are different from each other")
for i in data_cor.columns:
    print("*"*30)
    print(i)
    print("*"*30)
    x= str(i)
    a=data_cor[data_cor["{}".format(x)]<-0.8]
    
    print(a.index)
    print("=="*40)
```
