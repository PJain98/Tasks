# In this assignment we have to compress racoon grey scale image into  5 clusters. In the end, visualize both raw and compressed image and look  for quality difference. 


```python
import numpy as np 
from sklearn import cluster, datasets
from sklearn.cluster import KMeans
from scipy import misc 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face = misc.face(gray=True)
```

```python
plt.figure(figsize=(10, 3.5))
plt.imshow(face, cmap='gray')
plt.show()
```

```python
face.shape
```

```python
row = face.shape[0]
row
```

```python
col = face.shape[1]
col
```

```python
f = face.reshape(row*col,1)
```

```python
Kmean= KMeans( n_clusters=5,init="k-means++")
```

```python
Kmean.fit(f)
```

```python
y = Kmean.predict(f)
```

```python
y = y.reshape(row,col)
```

```python
y.shape
```

```python
plt.imshow(y)
plt.show()
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,5))
ax[0].imshow(face,cmap='gray')
ax[1].imshow(y)
```
